```python
# sklearn preprocessing for dealing with categorical variables
from sklearn.preprocessing import LabelEncoder
```

```python
#Encoding categorical variables <= 2 categories
le= LabelEncoder()
le_count = 0
for col in train:
    if train[col].dtype == 'object':
        if len(list(train[col].unique())) <=2:
            le.fit(train[col])
            train[col] = le.transform(train[col])
            test[col] = le.transform(test[col])
            le_count+=1
print('%d columns were label encoded.' % le_count)
```

```python
#One hot code for > 2 categories
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

Aligning label between Train and test columns since there are a difference between them

```python
y_train  = train['TARGET']

# Align the training and testing data, keep only columns present in both dataframes
train, test = train.align(test, join = 'inner', axis = 1)

train['TARGET'] = y_train
print('Training Features shape: ', train.shape)
print('Testing Features shape: ', test.shape)
```

```python
# Age information into a separate dataframe
age_data = train[['TARGET', 'DAYS_BIRTH']]

# Calculate size of bins dynamically and assign range as labels
num_bins = 11  # Number of desired bins

# Calculate bin edges based on quantiles
quantiles = np.round(pd.qcut(train['DAYS_BIRTH'], q=num_bins, retbins=True)[1],1)

# Assign range labels to bins
labels = [f'{quantiles[i]}-{quantiles[i+1]}' for i in range(num_bins-1)]
labels.append(f'>{quantiles[-2]}')
age_data['Age Group'] = pd.cut(train['DAYS_BIRTH'], bins=quantiles, labels=labels, right=False)
```

Using Random Forest Tree<br>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split,GridSearchCV
```

Preprocessing data before training ML

```python
# Separate the features and the target variable
X = train.drop(['TARGET','SK_ID_CURR'], axis=1)
y = train['TARGET']


#Copy dataset
# X_train = X.copy()
# X_test = test.copy()
test = test.drop(['SK_ID_CURR'], axis=1)

# Median imputation of missing values
imputer = SimpleImputer(strategy = 'median')

# Scale each feature to 0-1
scaler = MinMaxScaler(feature_range = (0, 1))

# Fit on the training data
imputer.fit(X)

# Transform both training and testing data
X = imputer.transform(X)
test = imputer.transform(test)

# Repeat with the scaler
scaler.fit(X)
X = scaler.transform(X)
test = scaler.transform(test)

print('Training data shape: ', X.shape)
print('Testing data shape: ', test.shape)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Make the random forest classifier
rf_classifier = RandomForestClassifier(n_estimators = 100, random_state = 42)
#Classfication problem sqrt(p) 

rf_classifier.fit(X_train, y_train)
```

```python
y_pred = rf_classifier.predict(X_test)
# # Submission dataframe
# submit = test[['SK_ID_CURR']]
# submit['TARGET'] = y_pred

# submit.head()
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Generate a classification report
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```

```python
# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

Evaluate the model using ROC and AUC

```python
# Predict probabilities
y_pred_proba = rf_classifier.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
# Calculate false positive rate, true positive rate, and thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calculate AUC score
auc_score = roc_auc_score(y_test, y_pred_proba)

# Print AUC score
print("AUC Score:", auc_score)
```

```python
# Plot ROC curve
plt.plot(fpr, tpr, label='ROC Curve (AUC = %0.2f)' % auc_score)
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random classifier
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()
```

```python
# Get feature importances
importance = rf_classifier.feature_importances_
```

```python
#Copy dataset
train_copy = train.copy()
train_copy = train_copy.drop(['TARGET','SK_ID_CURR'], axis=1)
```

```python
# Create a DataFrame to store feature importances
feature_importances = pd.DataFrame({'Feature': train_copy.columns, 'Importance': importance})
feature_importances = feature_importances.sort_values('Importance', ascending=False)
```

```python
# Plot top 20 feature importances
top_20_features = feature_importances.head(20)
plt.figure(figsize=(10, 6))
plt.barh(top_20_features['Feature'], top_20_features['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Top 20 Feature Importances')
plt.show()
```

```python
top_10_features = feature_importances.head(10)['Feature']
X_top_10 = train_copy[top_10_features]
correlation_matrix = X_top_10.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Top 10 Features)')
plt.show()
```

```python
# Define the hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(rf_classifier, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters and score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Print the best hyperparameters and score
print("Best Hyperparameters:", best_params)
print("Best Score:", best_score)
```
